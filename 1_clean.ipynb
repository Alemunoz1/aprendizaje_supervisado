```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Suponiendo que tu dataframe se llama 'df'
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')

plt.title('Mapa de Calor de Valores Nulos')

# Guardar el gráfico como un archivo PNG
plt.savefig('output.png', format='png')

# Mostrar el gráfico
plt.show()
```

```python
missing_values_percentage = df.isnull().mean() * 100
missing_values_percentage = missing_values_percentage.round(2).sort_values(ascending=False)
print(missing_values_percentage)
```

```python
# Importamos librerías
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Contar los valores faltantes por columna
missing_counts = df.isnull().sum().sort_values(ascending=False)

# Crear un DataFrame para seaborn
missing_df = missing_counts.reset_index()
missing_df.columns = ['Columna', 'Valores faltantes']

# Crear la gráfica de barras
plt.figure(figsize=(10,6))
sns.barplot(x='Columna', y='Valores faltantes', data=missing_df, palette='viridis')
plt.xticks(rotation=45)
plt.title('Valores faltantes por columna')
plt.show()
```

```python
df.drop('Cabin', axis = 1, inplace = True)
```

```python
# Calcular la media
mean_age = df['Age'].mean()

# Imputar los valores nulos reemplazando la columna directamente
df['Age'] = df['Age'].fillna(mean_age)
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
round(df.isnull().sum().sort_values(ascending = False) / len(df) * 100, 2)
```
